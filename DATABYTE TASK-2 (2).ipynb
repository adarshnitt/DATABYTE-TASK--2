```python
data_set="iris.csv"
def data_prep(data_set=data_set):
    import pandas as pd
    import seaborn as sns
    data=pd.read_csv(data_set)
    y=data["species"]
    x=data.drop(["species"],axis=1)
    return x,y

import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(14)
plt.subplot(1,2,1)
sns.scatterplot(x=x["sepal_length"],y=x["sepal_width"])
plt.subplot(1,2,2)
sns.scatterplot(x=x["petal_length"],y=x["petal_width"])
plt.show()
sns.kdeplot(data=x)
plt.show()
```

```python
def model_knn(sepal_length="sepal_length",sepal_width="sepal_width",
              petal_length="petal_length",petal_width="petal_width"):
    
    
    from sklearn.neighbors import KNeighborsClassifier
    data_prep()
    model=KNeighborsClassifier()
    model.fit(x,y)
    value=[sepal_length,sepal_width,petal_length,petal_width]
    predict=model.predict([value])
    print("given flowe is ",predict )
```

```python
def model_LR(sepal_length="sepal_length",sepal_width="sepal_width",
             petal_length="petal_length",petal_width="petal_width"):
    
    from sklearn.ensemble import AdaBoostClassifier
    from sklearn.linear_model import LogisticRegression
    data_prep()
    model_lr=LogisticRegression()
    ada=AdaBoostClassifier(base_estimator=model_lr,n_estimators=500,learning_rate=1)
    ada.fit(x,y)
    value=[[sepal_length,sepal_width,petal_length,petal_width]]
    predict=ada.predict(value)
    print("given flower is ",predict) 
```

```python
model_LR(sepal_length=5.9,sepal_width=3.0,
              petal_length=5.11,petal_width=1.8)
```

```python
model_knn(sepal_length=5.9,sepal_width=3.0,
              petal_length=5.11,petal_width=1.8)
```
